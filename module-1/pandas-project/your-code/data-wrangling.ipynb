```python
# ordenar by species 
df_notnulls.rename(columns={'Species ': 'Species'}, inplace = True)
```

```python
#nueva columna especie 
df_notnulls['Species_clean'] = df_notnulls['Species'].copy()
```

```python
#palabras que eliminar de la columna 

words = ["Possibly","for","no","fragments","larger","less","than","by",
        "determined","adult","male","hooked","captured","involved","cm",
        "feet","female","tooth", "fragment", "recovered","according", 
        "witnesses","said","be","averaged","length","pack","of","seen", 
        "in", "vicinity",'',"kg","lb","small","a","m","thought", "to", 
        "involve","possibly", "juvenile","suspected", "due", "freshwater", 
        "habitat"]


def species_classification(x):
    if x['Species'] == x['Species']:
        especies_div = x['Species'].split()
        especie_stop = []
        for i in especies_div:
            x = i.lower()
            x = re.sub('[^a-zA-Z]', "", x)
            x = x.replace('sharks', 'shark')
            if x not in words:
                especie_stop.append(x.lower())
        especie = ' '.join(especie_stop).capitalize()
        if "shark" in especie_stop:
            i_shark = especie_stop.index('shark')
            if len(especie_stop)>2:
                especie_alt = (especie_stop[i_shark-2]+''+ especie_stop[i_shark-1]+ ' ' + especie_stop[i_shark]).capitalize()
            else:
                especie_alt = (especie_stop[i_shark-1] + ' ' + especie_stop[i_shark]).capitalize()
        else: 
            
            especie_alt = 'other species'
            
            
        if len(especie) == 0:
            
            return "not confirmed"
        elif re.search('[Nn]ot', especie) or re.search('.*un.*', especie) or re.search('.*[Qq]ue.*', especie) or re.search('[Uu]ni', especie) or especie == 'Shark':
    
    
            return 'Not confirmed'
        elif re.search('[Ii]nvalid', especie) or re.search('No shark involment', especie):
        
            return 'No shark involvment'
        elif re.search('[Ww]ite\s', especie):
            return 'White shark'
        elif re.search('[Tt]ig', especie):
            return 'Tiger shark'
        elif re.search('[Ss]pin', especie):
            return 'Spinner shark'
        else:
            return especie_alt
    else:
        
        return 'Not confirmed'
    
```

```python
df_notnulls['Species_clean'] = df_notnulls.apply(lambda x: species_classification(x), axis = 1)
```

```python
df_notnulls['Age_classified'] = df_notnulls['Age'].copy()
```

```python
def age_classification(x):
    if x['Age'] == x['Age']:
        age = x['Age']
        
        #se quita lo que no sean numeros
        age = re.sub('[^0-9]', '', age)
        age = list(map(int, age.split()))
        
        #verificamos cuantos numeros hay en la edad
        if len(age) == 0:
            return 'no record'
        elif len(age) == 1:
            age = age[0]
        else: 
            age = round(sum(age)/ len(age))
            
        # clasificacion de edad en rangos
        
        if age <= 10:
            return '0 - 10'
        elif age <= 20:
            return '11-20'
        elif age <= 30:
            return '21-30'
        elif age <= 40:
            return '31-40'
        elif age <= 50:
            return '41-50'
        elif age <= 60:
            return '51-60'
        elif age <= 70:
            return '61-70'
        elif age <= 80:
            return '71-80'
        else: 
            return '81 - more'
    else:
        return 'no record'
            
        
            
```

```python
df_notnulls['Age_classified'] = df_notnulls.apply(lambda x: age_classification(x), axis= 1)
```

```python
df_notnulls['Year_classified'] = df_notnulls['Year'].copy()
```

```python
def year_classifier(x):
    if x['Year'] == x['Year']:
        year = int(x['Year'])
        if len(str(year)) < 4:
            return 'no record'
        else:
            return year
    else:
        return 'no record'
    
    
    
        
    
```

```python
df_notnulls['Year_classified'] = df_notnulls.apply(lambda x: year_classifier(x), axis = 1)
```

```python
df_notnulls['Month_classifier'] = df_notnulls['Date'].copy()
```

```python
# funcion para limpiar columna de meses 

def month_classifier(x):
    months_dict = { '[Jj]an': 'January',
                    '[Ff]eb': 'February',
                   '[Mm]ar' : 'March',
                   '[Aa]pr' : 'April',
                   '[Mm]ay' : 'May',
                   '[Jj]un' : 'June',
                   '[Jj]ul' : 'July',
                   '[Aa]ug' : 'August',
                   '[Ss]ep' : 'September',
                   '[Oo]ct' : 'October',
                   '[Nn]ov' : 'November',
                   '[Dd]ec' : 'December'
                  }
    if x['Date'] == x['Date']:
        month = x['Date']
        for key, value in months_dict.items():
            if re.search(key, month):
                return value
            
        return 'no record'
    
    else:
        return 'no record'
    
    
                
            
    
```

```python
df_notnulls['Month_classifier'] = df_notnulls.apply(lambda x: month_classifier(x),
                                                   axis = 1)

df_notnulls['Month_classifier'].value_counts()
```

```python
#Clasificador tipos de ataque

df_notnulls['Type_classified'] = df_notnulls['Type'].copy()

def type_classifier(x):
    if x['Type'] == x['Type']:
        types = x['Type']
        if re.search('[Bb]oat', types):
            return 'Something in a boat activitie'
        else:
            return types
    else:
        return 'Not especified'
    

# aplicamos funcion en columna type

df_notnulls['Type_classified'] = df_notnulls.apply(lambda x: type_classifier(x),
                                                  axis = 1)

#nueva columna type
df_notnulls['Type_classified'].value_counts()    
```

```python
##limpieza de country 

df_notnulls['Country_classified'] = df_notnulls['Country'].copy()


no_words = [ "between" , "federated", "states", "of", "bay", "coast", "overseas" , "territory", "territories" ]
            
# funcion para limpiar
            
def country_classifier(x):
            if x['Country'] == x['Country']:
                country = x['Country'].strip().lower()
                country = re.sub('[^ a-zA-Z &\\/]', '', country)
                if '/' in country:
                    country = country.split('/')
                    country = country[0]
                country_lst = country.split()
                country_stop = []
                for i in country_lst:
                    if i not in no_words:
                        country_stop.append(i.capitalize())
                country = ' '.join(country_stop)
                return country_verification(country)
            else: 
                return 'Not specified'
            
# funcion regex en paises 

def country_verification(country):
            countries_dict = {'[Oo]cean': 'Not specified',
                              '[Ss]ea' : 'Not specified',
                              '^[Aa]frica$': 'Country in Africa',
                              '[Aa]ndaman' : 'Andaman Islands',
                              '[Cc]eylon' : 'Sri Lanka',
                              '[Aa]sia' : 'Country in Asia'
                                }
            for key, value in countries_dict.items():
                if re.search(key, country):
                    return value
            return country
                
            
#aplicamos funciones                
df_notnulls['Country_classified'] = df_notnulls.apply(lambda x: country_classifier(x)
                                                     ,axis = 1)

df_notnulls['Country_classified'].value_counts()
            
      
```

```python
# cleaning activity

df_notnulls['Activity_classified'] = df_notnulls['Activity'].copy()

def activity_classification(x):
    if x['Activity'] == x['Activity']:
        activity = x['Activity'].strip().lower()
        activity = re.sub('[^a-zA-Z \\/]', '', activity)
        activity = activity.split()
        verbs = []
        
        for i in activity:
            txt = re.findall('.*ing', i)
            if len(txt) >=1:
                verbs.append(txt[0])
        if len(verbs)>=1:
            if 'shark' in activity or 'sharks' in activity:
                act = verbs[0]+' sharks'
                return act.capitalize()
            else: 
                return verbs[0].capitalize()
        else:
            
            return 'not specified'
    else: 
        return 'not specified'

    
df_notnulls['Activity_classified'] = df_notnulls.apply(lambda x: activity_classification(x),
                                                      axis = 1)

df_notnulls['Activity_classified'].value_counts()

    
    
```

```python
#Se arregla nombre acolumna Sex

df_notnulls.rename(columns={'Sex ' : 'Sex'}, inplace = True)

df_notnulls['Sex'] = df_notnulls['Sex'].copy()

def sex_classification(x):
    if x['Sex'] == x['Sex']:
        gender = x['Sex']
        if re.search('[Mm]', gender):
            return 'M'
        elif re.search('[Ff]', gender):
            return 'F'
        else:
            return 'No specified'
    else:
        return 'No specified'
    
#Limpiamos genero con funcion

df_notnulls['Sex_classified'] = df_notnulls.apply(lambda x: sex_classification(x), axis = 1)

#Final

df_notnulls['Sex_classified'].value_counts()
```

```python
#se crea columna de fatal limpia
df_notnulls['Fatal_classified'] = df_notnulls['Fatal (Y/N)'].copy()

#funcion para limpiar columna fatal
def fatal_classification(x):
    if x['Fatal (Y/N)'] == x['Fatal (Y/N)']:
        
        #fatal = x['Fatal (Y/N)']
        if re.search('[Yy]', x['Fatal (Y/N)']):
            return 'Yes'
        elif re.search('[N/n]', x['Fatal (Y/N)']):
            return 'No'
        else:
            return 'not specified'
    else:
        return 'not specified'
    
    
#usamos funcion

df_notnulls['Fatal_classified'] = df_notnulls.apply(lambda x: fatal_classification(x), axis = 1 )

df_notnulls['Fatal_classified'].value_counts(normalize = True)

###insight 

```

```python
#lipieaza de injury
df_notnulls['Injury'].values



df_notnulls['Injury_classified'] = df_notnulls['Injury'].copy()

def injury_classification(x):
    if x['Injury'] == x['Injury']:
        injury = x['Injury'].strip().lower()
        injury = re.sub('[^a-zA-Z]', '', injury)
        return injury_verification(injury)
    else: 
        return 'Not specified'
    
def injury_verification(injury):
    injury_dict = {'fatal':'Fatal', 'no injury':'No injury',
                  'bitten|right|left|injur|bite':'Body bite / injured',
                  'no.*details|surv|not|nonfatal':'Not specified',
                  'lacera|abra':'Body lacerations',
                  'severed|ampu':'Amputed limb', 
                  'minor|punc':'Minur injury',
                  'drown':'Drowned'}
    for key, value in injury_dict.items():
        if re.search(key, injury):
            return value
    return 'Not specified'

df_notnulls['Injury_classified'] = df_notnulls.apply(lambda x: injury_classification(x), axis = 1)


df_notnulls['Injury_classified'].value_counts()
    

```

```python
#clean Name 

df_notnulls['Name_classified'] = df_notnulls['Name'].copy()

stop = ["yacht", "owened", "by","launch","a","pilot","air","force","officer","royal","australian","skiff", "occupant","captain","occupants"]
name_dict = {"valet|atla|coa|dingh|teen|peo|passe|teac|male|boy|boat|child|anony|sailor|girl|unid|unkn|fish|div|native|sold|men$|wom|young|yout|crew":"Not confirmed"}


def name_classification(x):
    
    
    if x['Name'] == x['Name']:
        name = x['Name'].lower()
        name_lst = name.split(',')
        name_lst = name_lst[0].strip()
        name_lst = name_lst.split()
        name_stop = []
        for i in name_lst:
            if i not in stop:
                name_stop.append(i.capitalize())
        if len(name_stop) < 2:
            return 'Not confirmed'
            
        else: 
                name = ' '.join(name_stop)
                name = re.sub('[^a-zA-z .]', "", name)
        for key, value in name_dict.items():
                if re.search(key, name.lower()):
                    return value
            
        if len(name.split())< 2:
                return 'Not confirmed'
        else: 
                return name
        
    else:
        return 'Not confirmed'
                

df_notnulls['Name classified'] = df_notnulls.apply(lambda x: name_classification(x), axis = 1)
df_notnulls['Name classified'].value_counts()
```
