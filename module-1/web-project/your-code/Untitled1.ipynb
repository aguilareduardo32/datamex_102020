```python


from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd

url = 'https://www.marketwatch.com/story/this-timeline-charts-elon-musks-dramatic-rise-to-internet-fame-or-descent-into-internet-infamy-2018-08-31'
page = requests.get(url)

#page



soup = BeautifulSoup(page.content, 'html.parser')
soup


results = soup.find_all('p')

titles = soup.find_all('strong')

#print(results)



# se busca scrapear el texto con las fechas para analizar ó
#todos los textos que tengan <strong> que es la fecha despues tienen un <p> que es la descripcion de tweet


description = []

for elem in titles:
    #print(elem.findNext('p').text)
    description.append(elem.findNext('p').text)
    
description = description[2:]
description.pop(-1)
#print(description)



len(description) == len(titles)

#despues de encontrar la descripcion se limpiar la fecha que venia dentro de un tag

titles = [t.get_text() for t in soup.find_all('strong')]
titles = titles[2:-1]
titles

description = [re.sub('\n', '', i) for i in description]
#print(description)

tweets_list = pd.DataFrame(
    { 'Tweet Date': titles,
     'Tweet descrption': description
    })


tweets_list.to_csv (r'tweets_dataframe.csv', index = False, header=True)

#removing tags from dates re

print(tweets_list['Tweet descrption'][33])
print(tweets_list['Tweet descrption'][15])
print(tweets_list['Tweet descrption'][24])
tweets_list


# el 28 hubo mucho cambio por un asunto de Elon con la SEC pero el no tweetio nada por eso no se analizo
#Se hizo scrap de la descripcion y no del tweet por que no se puede hacer scrap de una api de twitter ni de twitter,(se intento)
```
