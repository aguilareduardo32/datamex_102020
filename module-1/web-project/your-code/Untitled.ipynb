```python
start_volume = datetime.date(2018, 9, 4)
end_volume = datetime.date(2018, 9, 15)

df_volume = get_historical_data('TSLA', start_volume, end_volume, output_format='pandas', token= api_key)
df_volume['volume']


#se agrega columna CHANGE de cambio de precio en cada dia 
df_volume['Change'] = df_volume['open'].pct_change()
df_volume['Change'] = [i * 100 for i in df_volume['Change']]
#print(df_volume)

## sacar el promedio de cuanto cambia en ese mes  y la diferencia de cuanto cambio ese dia en precio 

start_month_change = datetime.date(2018, 9, 1)
end_month_change = datetime.date(2018, 9, 30)

df_month_change = get_historical_data('TSLA', start_month_change, end_month_change, output_format='pandas', token= api_key)




df_month_change['Price Change'] = df_month_change['open'].pct_change()
df_month_change['Price Change'] = [i * 100 for i in df_month_change['Price Change']]
df_month_change

cambio_mensual = df_month_change['Price Change'].sum() / len(df_month_change['Price Change'])
print(f'Cambio promedio mensualmente en precio de la accion TSLA en el mes de Septiembre: {cambio_mensual}')
cambio_Joe = df_month_change['Price Change'][3]

print(f'El dia despues de Joe Rogan experience el precio cambio: {cambio_Joe}' )

# cambio porcentaje de cambio en volumen 

df_month_change['Volume change'] = df_month_change['volume'].pct_change()
df_month_change['Volume change'] = [i * 100 for i in df_month_change['Volume change']]


## TALVEZ EL cambio del valor de la compania completa 
print('Se puede ver que los dias con mas cambio en el volumen fueron los dias que el precio de la accion tuvo mayores disminuciones debido a las leyes de oferta y demanda')
df_month_change

# el 28 hubo mucho cambio por un asunto de Elon con la SEC pero el no tweetio nada por eso no se analizo
```
